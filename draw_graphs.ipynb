```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


MAPk_vals = np.load('./data/MAPk_vals.npy')
cos_dissim_vals = np.load('./data/cos_dissim.npy')

# baseline = np.load('./data/baseline_performance.npy')
baseline = np.array([0.004,0.17])

knn_frac_vals = [0.1,0.3,0.5,0.7,0.9]
min_overlap_vals = [0, 0.05, 0.1,0.15,0.2]
k_vals = [10,50,100,150,200]


```

```python
MAPk_vals
```

```python
#kNN frac
plt.figure()
plt.title("MAP@K for varying MUSIC profile weight")
plt.xlabel("Fraction of songs determined by MUSIC profile")
plt.ylabel("MAP@K value")
plt.plot(knn_frac_vals,MAPk_vals[0,:])
plt.hlines(baseline[0],knn_frac_vals[0],knn_frac_vals[4],linestyles='dashed')
plt.legend(['Novel Method','Baseline'])
plt.savefig("./figures/mapk_knn_frac")

plt.figure()
plt.title("Cosine list dissimilarity for varying MUSIC profile weight")
plt.xlabel("Fraction of songs determined by MUSIC profile")
plt.ylabel("List Dissimilarity")
plt.plot(knn_frac_vals,cos_dissim_vals[0,:])
plt.hlines(baseline[1],knn_frac_vals[0],knn_frac_vals[4],linestyles='dashed')
plt.legend(['Novel Method','Baseline'])
plt.savefig("./figures/div_knn_frac")
```

```python
#overlap
plt.figure()
plt.title("MAP@K for varying minimum MUSIC profile overlap")
plt.xlabel("Minimum MUSIC profile overlap")
plt.ylabel("MAP@K value")
plt.plot(min_overlap_vals,MAPk_vals[1,:])
plt.hlines(baseline[0],min_overlap_vals[0],min_overlap_vals[4],linestyles='dashed')
plt.legend(['Novel Method','Baseline'])
plt.savefig("./figures/mapk_overlap")


plt.figure()
plt.title("Cosine list dissimilarity for varying minimum MUSIC profile overlap")
plt.xlabel("Minimum MUSIC profile overlap")
plt.ylabel("List Dissimilarity")
plt.plot(min_overlap_vals,cos_dissim_vals[1,:])
plt.hlines(baseline[1],min_overlap_vals[0],min_overlap_vals[4],linestyles='dashed')
plt.legend(['Novel Method','Baseline'])
plt.savefig("./figures/div_overlap")
```

```python
#k_vals
plt.figure()
plt.title("MAP@K for number of kNN neighbours")
plt.xlabel("Number of kNN neighbours")
plt.ylabel("MAP@K value")
plt.plot(k_vals,MAPk_vals[2,:])
plt.hlines(baseline[0],k_vals[0],k_vals[4],linestyles='dashed')
plt.legend(['Novel Method','Baseline'])
plt.savefig("./figures/mapk_knn")

plt.figure()
plt.title("Cosine list dissimilarity for varying number of kNN neighbours")
plt.xlabel("Number of kNN neighbours")
plt.ylabel("List Dissimilarity")
plt.plot(k_vals,cos_dissim_vals[2,:])
plt.hlines(baseline[1],k_vals[0],k_vals[4],linestyles='dashed')
plt.legend(['Novel Method','Baseline'])
plt.savefig("./figures/div_knn")
```
