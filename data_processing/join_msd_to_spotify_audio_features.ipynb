```python
audio_features = audio_features.drop(columns=['analysis_url', 'uri', 'type', 'track_href'])
audio_features.head()
```

```python
msd = msd.drop(columns=['danceability', 'duration', 'energy', 'key', 'loudness', 'mode', 'tempo', 'time_signature'])
msd.head()
```

```python
msd_with_audio_features = msd.merge(audio_features, left_on='spotify_id', right_on='id')
msd_with_audio_features = msd_with_audio_features.drop(columns=['id'])
msd_with_audio_features.head()
```
