```python
import pandas as pd
```

```python
# track_genres = pd.read_table('http://www.ifs.tuwien.ac.at/mir/msd/partitions/msd-MAGD-genreAssignment.cls',names=['TrackID','Genre'],sep='\t')
track_genres = pd.read_table(
    '../data/msd-MAGD-genreAssignment.cls',
    names=['track_id', 'genre'],
    sep='\t')
msd = pd.read_hdf('../data/full_msd.h5', key='df')

track_genres['track_id'] = track_genres['track_id'].str.encode('utf-8')
track_genres['genre'] = track_genres['genre'].str.encode('utf-8')
track_genres['track_id'] = track_genres['track_id'].astype(str)
track_genres['genre'] = track_genres['genre'].astype(str)
msd['track_id'] = msd['track_id'].astype(str)
```

```python
msd_with_genre = pd.merge(msd, track_genres,on='track_id', how='inner')
msd_with_genre.head()
```
