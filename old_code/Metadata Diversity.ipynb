# Metadata Diversity

Given a list of track_id's, the metadata diversity of the list is:

List_Diversity = (A)num_genres + (B)num_albums + (C)num_artists + (D)num_decades + ...
Where A,B,C,D,... are positive constants

```python
import pandas as pd
import numpy as np
```

```python
def metadata_diversity(track_ids):
    df = pd.read_csv('song_data.csv')
    df = df[df['track_id'].isin(track_ids)]
    
    diversity = 0
    diversity += 1*df['genre'].nunique() # genres
    diversity += 1*df['release_7digitalid'].nunique() # albums
    diversity += 1*df['artist_name'].nunique() # artists
    diversity += 1*df['year'].nunique() # years
    diversity += 1*np.floor(df['year']/10).nunique() # decades
    diversity += 1*np.round(df['duration']/60).nunique() # durations in minutes
    diversity += 1*np.round(df['tempo']/10).nunique() # tempo 10s of BMP
    
    return diversity
```
