```python
# calculate diversity of top n vectors

def get_diversity(song_vector_list):
    
    dissim = 0
    
    n = len(song_vector_list)
    
    for i in range(n):
        for j in range(n):
            dissim += np.linalg.norm(song_vector_list[i] - song_vector_list[j])
            
    return dissim/((n/2)*(n-1))
```

```python
songs_df = pd.read_hdf('data/full_msd_with_audio_features.h5', key='df')
songs_df.head()
print(list(songs_df.columns.values))
```

```python
def convert_recs_to_embeddings(rec_list, songs_df):
    embeddings_list = []
    
    for rec_ids in rec_list:
        #print(prepare_song_id(rec_ids[0]))
        if (prepare_song_id(rec_ids[0]) in songs_df.song_id.values):
            row = songs_df.loc[songs_df['song_id'] == prepare_song_id(rec_ids[0])]
            embedding = row.iloc[:,12:]
            embedding = np.array(embedding.values.tolist()[0])
            embeddings_list.append(embedding)
        
    return embeddings_list
```

```python
# test out diversity levels

recommendations = np.load('data/recommendations.npy')

# 1000 lists of the top 100 recommendations, featuring two identifying keys for each recommendation

diversities = []
i = 0
for rec_list in recommendations:
    
    # convert to list of 100 embeddings
    embedding_list = convert_recs_to_embeddings(rec_list, songs_df)
    
    # calculate diversity
    diversities.append(get_diversity(embedding_list))
    
    print(i)
    i += 1
    
print(diversities.mean)
print(diversities.max)
print(diversities.min)
```
